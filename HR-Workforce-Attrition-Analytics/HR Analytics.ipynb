#### ==============================
### 4. OVERALL KPI ANALYSIS
#### ==============================

### 4.1 Total Employees

```python
total_employees = len(df)
print("Total Employees:", total_employees)
```

### 4.2 Active Employees

```python
active_employees = df[df["Attrition"]=="No"].shape[0]
print("Active Employees:", active_employees)
```

### 4.3 Employees Left

```python
employees_left = df[df["Attrition"] == "Yes"].shape[0]
print("Employees Left:", employees_left)
```

### 4.4 Attrition Rate

```python
attrition_rate = round(
      (employees_left/total_employees)*100,
    2)
attrition_rate
print("Atrition Rate:", attrition_rate, "%")
```

### 4.5 KPI Summary

```python
summary = pd.DataFrame({
             "Metric":[
                       "total_employees",
                       "active_employees",
                       "employees_left",
                       "attrition_rate"
                      ],
             "Values":[total_employees,
                       active_employees,
                       employees_left,
                       attrition_rate
                      ]
})                
summary
```

### Insight:
#### The organization has 1,470 employees, of which 1,233 are active and 237 have left the company, resulting in an overall attrition rate of 16.12%.

#### ========================
### 5. DEPARTMENT ANALYSIS
#### ========================

### 5.1 Attrition by Department

```python
department_attrition = (
 df[df["Attrition"] == "Yes"]
.groupby("Department")
.size()
.reset_index(name="Employees_Left")
.sort_values(by="Employees_Left", ascending = False)
 )

department_attrition
```

### 5.2 Department-wise Attrition Rate

```python
department_attrition_rate = (
    df.groupby("Department")["Attrition"]
    .apply(lambda x: round((x == "Yes").mean() * 100, 2))
    .reset_index(name="Attrition_Rate")
    .sort_values(by="Attrition_Rate", ascending=False)
)

department_attrition_rate
```

### 5.3 Average Salary by Department

```python
department_salary = (df.groupby("Department")["MonthlyIncome"]
.mean()
.round(2)
.reset_index(name="Average_Salary")
.sort_values(by="Average_Salary", ascending = False)
                    )                  
department_salary
```

### Insight:
#### Research & Development recorded the highest number of employees leaving (133), 
#### while Sales had the highest attrition rate (20.63%) and the highest average monthly salary (6,959.17) among all departments.

#### ====================
### 6. OVERTIME ANALYSIS
#### ====================

 ### 6.1 Attrition by Overtime

```python
overtime_attrition = (
df[df["Attrition"]=="Yes"]
.groupby("OverTime")
.size()
.reset_index(name="Employees_Left")
.sort_values(by="Employees_Left", ascending = False)
)
overtime_attrition
```

### Insight:
#### Employees who worked overtime had a higher number of resignations (127) compared to employees who did not work overtime (110), indicating that overtime may be associated with increased employee attrition.

#### ====================
### 7. AGE GROUP ANALYSIS
#### ====================

### 7.1 Create Age Group

```python
df["Age_Group"] = np.where(
    df["Age"] <= 30, "Young",
    np.where(df["Age"] <= 50, "Adult", "Senior")
)

df[["Age", "Age_Group"]].head()
```

### 7.2 Attrition by Age Group

```python
age_group_attrition = (
df[df["Attrition"]=="Yes"]
.groupby("Age_Group")
.size()
.reset_index(name="Employees_Left")
.sort_values(by="Employees_Left",ascending = False)
)
age_group_attrition
```

### Insight:
#### Adult employees had the highest attrition (119), followed by Young employees (100), while Senior employees had the lowest attrition (18).

#### ====================
### 8. JOB ROLE ANALYSIS
#### ====================

### 8.1 Attrition by Job Role

```python
job_role_attrition = (
 df[df["Attrition"]=="Yes"]
.groupby("JobRole")
.size()
.reset_index(name= "Employees_Left")
.sort_values(by="Employees_Left",ascending=False)
)
job_role_attrition
```

### Insight:
#### Laboratory Technicians had the highest attrition (62), followed by Sales Executives (57) and Research Scientists (47), while Research Directors had the lowest attrition (2).

### 9. EMPLOYEES EARNING MORE THAN COMPANY AVERAGE SALARY

```python
above_average_salary = (
    df[df["MonthlyIncome"] > df["MonthlyIncome"].mean()]
    .sort_values("MonthlyIncome",ascending=False)
)
above_average_salary
```

### Insight:
#### A total of 493 employees earn more than the company's average monthly salary, helping identify high-income employees across different departments and job roles.

### 10. SALARY RANKING WITHIN EACH DEPARTMENT

```python
salary_rank = (
    df.assign(
        Department_Salary_Rank=df.groupby("Department")["MonthlyIncome"]
        .rank(method="dense", ascending=False)
    )
    .sort_values(by=["Department", "Department_Salary_Rank"])
)

salary_rank
```

### Insight:
#### Employees are ranked by monthly salary within each department, with employees earning the same salary receiving the same rank using the dense ranking method.

### 11. TOP-PAID EMPLOYEE IN EACH DEPARTMENT

```python
Top_Paid_Employee = (
 df.sort_values(by="MonthlyIncome", ascending = False)
.groupby("Department")
.head(1)
.reset_index(drop=True)
)
Top_Paid_Employee        
```

### Insight:
#### The analysis identifies the highest-paid employee in each department, showing that Research & Development had the highest monthly salary (19,999), followed by Sales (19,847) and Human Resources (19,717).
